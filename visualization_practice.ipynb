```python
fig = plt.figure()
a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)

x = range(0,10)
y = range(5, 15)
a1.plot(x,y, 'm+')
x = range(20)
y = range(20)
a2.plot(x,y, 'c.')

plt.show()
```

```python
import matplotlib
# 간격을 가능한 최적의 수치로 자동 조정 => constrained_layout=True
fig = plt.figure(constrained_layout=True)
a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)

x = range(0,10)
y = range(5, 15)
a1.plot(x,y, 'm+')
x = range(20)
y = range(20)
a2.plot(x,y, 'c.')

# 범례(x축, y축 이름 지정)
a1.set_xlabel('time')
a1.set_ylabel('money')
a2.set_xlabel('time')
a2.set_ylabel('age')

plt.show()
```

```python
from cProfile import label

# 막대그래프에 수치 표현
fig = plt.figure()
plt.style.use('ggplot')
g = fig.add_subplot(111)
x = [3,4]
y = [1,2]

# 제목
labels = ['English', 'mathmatics']
# 색깔
colors = ['red','blue']

# 막대그래프 => bar / barh
g.bar(x,y, tick_label = labels, color = colors)

for i, v in enumerate(y):
    g.text(v+2, i+1.05, str(v))
plt.show()
```
