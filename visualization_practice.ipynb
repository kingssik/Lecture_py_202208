```python
import matplotlib
import matplotlib.font_manager as fm
# 간격을 가능한 최적의 수치로 자동 조정 => constrained_layout=True
fig = plt.figure(constrained_layout=True)
# 폰트지정
font_location = "C:\Windows\Fonts\H2PORM.TTF"
fontname = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font',family = fontname)

a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)

x = range(0,10)
y = range(5, 15)
a1.plot(x,y, 'm+')
x = range(20)
y = range(20)
a2.plot(x,y, 'c.')

# 범례(x축, y축 이름 지정)
a1.set_xlabel('time')
a1.set_ylabel('money')
a2.set_xlabel('time')
a2.set_ylabel('age')

plt.show()
```

```python
from cProfile import label

# 막대그래프에 수치 표현
fig = plt.figure()
plt.style.use('ggplot')
g = fig.add_subplot(111)
x = [3,4]
y = [1,2]

# 제목
labels = ['English', 'mathmatics']
# 색깔
colors = ['red','blue']

# 막대그래프 => bar / barh
g.bar(x,y, tick_label = labels, color = colors)

for i, v in enumerate(y):
    g.text(v+2, i+1.05, str(v))
plt.show()
```

```python
name = ['hong', 'hong', 'lim']
age = [23,22,50]
home = ['seoul', 'busan', 'daejeon']

# 딕셔너리처럼 만드는 법
member3 = pd.DataFrame()
member3['name'] = name
member3['age'] = age
member3['home'] = home
member3

# 이름 컬럼만 출력하려면
# print(member3['name'])

# 데이터프레임에 성별 컬럼 추가하려면
gender = pd.Series(['M', 'F', 'M'], name='gender')
member3['gender'] = gender
print(member3)

# 집계함수
# 최솟값
print(member3["age"].min())

# 최댓값
print(member3["age"].max())

# 평균값
print(member3["age"].mean())

# 데이터프레임의 기본 통계 내용 출력
print(member3.describe())

# head(n) : 끊어오려는 상위 n개의 행의 수
print(member3.head(2))
# tail(n) : 끊어오려는 하위 n개의 행의 수
print(member3.tail(1))
```
